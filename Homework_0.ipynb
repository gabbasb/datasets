```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import datasets

cancer = datasets.load_breast_cancer()
print(cancer.keys())
print(cancer.feature_names)
print(cancer.target_names)
print(cancer.DESCR)

malignant_data_list = np.where(cancer.target == 0)
print("There are ",len(malignant_data_list[0])," malignant cases.")


df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
display(df.head())
display(df.shape)

print('All columns')
print(list(df))





print(df['mean radius'])

print('display top 10 rows')
display(df[:10])

print('display 100th row')
display(df[99:100])

print('row at location 98')
print(df.loc[98])
print('row at imaginary location 99')
print(df.iloc[99])



# Add 'diagnosis' key for convenience 
df['diagnosis'] = np.array([cancer.target_names[i] for i in cancer.target])
display(df['diagnosis'])

# Plot a scatter plot of 'mean compactness' and 'mean perimeter'
sns.scatterplot(x='mean compactness', y='mean perimeter', hue='diagnosis', data=df)


sns.pairplot(df, vars=['mean texture', 'mean perimeter', 'mean compactness', 'mean radius'], hue='diagnosis')

#sns.pairplot(data=df)
```
